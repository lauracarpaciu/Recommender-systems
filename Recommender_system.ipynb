```python
def get_movie_rating(df, movieIds):
    ratings = []
    
    for movieId in movieIds:
        rating = df.query('@movieId == movieId')['rating'].mean()
        ratings.append(rating)
          
    return ratings 
```

```python
def get_most_common_rating(df, movieIds):
    ratings = []
    
    for movieId in movieIds:
        rating = df.query('@movieId == movieId')['rating'].value_counts()
        rating = rating.idxmax() if len(rating) > 0 else None
        ratings.append(rating)
          
    return ratings
```

```python
movies_id_list = movies_df['movieId'].unique()
movie_popularity_df = pd.DataFrame(index = movies_id_list);
movie_popularity_df['movieId'] = movies_id_list
```

```python
ratings = get_most_common_rating(ratings_df, movies_id_list)
```

```python
movie_popularity_df['rating'] = ratings
```
